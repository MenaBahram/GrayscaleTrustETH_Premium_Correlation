```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import pandas_datareader.data as data
import datetime 
import pylab

#Pull Assets
tickers = ["ETH-USD", "ETCG"]

#Set Time Frame
start = "2018-02-02"
end = datetime.date.today()

#Pull Data from Yahoo Finance
dfeth = data.DataReader("ETH-USD", data_source="yahoo", start=start, end=end)
# Find Premium
print(dfeth["Adj Close"])

#Pull Data from Yahoo Finance
dfet = data.DataReader("ETCG", data_source="yahoo", start=start, end=end)
print(dfet["Adj Close"])

#Premium
p = dfet["Adj Close"] / ((dfeth["Adj Close"]) * .001) - 1
print (p)


#Line Graph
plt.plot(dfet["Adj Close"])
plt.plot(dfeth["Adj Close"])
plt.plot(p)
#plt.yscale('log')
plt.xlabel("Date")
```

```python
# Correlation Matrix
df = pd.DataFrame({"Price GETH": dfet["Adj Close"],  "Price ETH": dfeth["Adj Close"],"Volume GETH": dfet["Volume"], "Volume ETH": dfeth["Volume"], "Premium": p})
corr = df.corr()
print()
corr.style.background_gradient(cmap='coolwarm')
```
